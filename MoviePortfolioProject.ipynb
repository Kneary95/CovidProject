```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\KevinNeary\OneDrive - Matrix Behavior Solutions, LLC\Documents\Python\movies.csv')
```

```python
# Looking at Data

df.head()
```

```python
# Checking For Missing Data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Dropping Missing Data
df = df.dropna()
```

```python
# Creataing Correct Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
df.head()
```

```python
# Dropping Any Duplicates
df.head().drop_duplicates()
```

```python
# Building Scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

```python
#Creating Regression Plot

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Showing Correlation
df.corr(numeric_only='TRUE')
```

```python
Correlation_Matrix = df.corr(numeric_only='TRUE')

sns.heatmap(Correlation_Matrix.round(2), annot=True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
# Looking at Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
Correlation_Matrix = df_numerized.corr(numeric_only='TRUE')

sns.heatmap(Correlation_Matrix, annot=True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Votes & Budget have highest correlation to gross earnings
# Company has lower correlation

High_Correlation = sorted_pairs[(sorted_pairs) > 0.5]

High_Correlation
```
